```python
from google.colab import files
import pandas as pd
import io
```

```python
# Função para carregar o arquivo CSV
def carregar_csv():
    uploaded = files.upload()
    for nome_arquivo in uploaded.keys():
        return pd.read_csv(io.BytesIO(uploaded[nome_arquivo]))

# Carregar o arquivo CSV pela primeira vez
df = carregar_csv()

# Exibir as primeiras linhas do dataset
print(df.head())
```

```python
class Perceptron:
    def __init__(self, n_atributos):
        self.n_atributos = n_atributos
        self.pesos = [0.0 for _ in range(n_atributos)]
        self.vies = 0.0

    def avanco(self, x):
        soma_ponderada = self.vies
        for i in range(self.n_atributos):
            soma_ponderada += x[i] * self.pesos[i]
        return 1 if soma_ponderada > 0 else 0

    def atualizacao(self, x, y_verdadeiro):
        previsao = self.avanco(x)
        erro = y_verdadeiro - previsao
        self.vies += erro  # Atualiza o viés
        for i in range(self.n_atributos):
            self.pesos[i] += erro * x[i]  # Atualiza os pesos
        return erro
```

```python
def treinamento(modelo, todos_os_x, todos_os_y, epocas):
    print("# INICIANDO O TREINAMENTO #")
    for epoca in range(epocas):
        contagem_de_erro = 0
        for x, y in zip(todos_os_x, todos_os_y):
            erro = modelo.atualizacao(x, y)
            contagem_de_erro += abs(erro)
        print(f'EPOCH {epoca + 1}, ERROS: {contagem_de_erro}')
        if contagem_de_erro == 0:
            break

def computar_acuracia(modelo, todos_os_x, todos_os_y):
    correto = 0
    for x, y in zip(todos_os_x, todos_os_y):
        previsao = modelo.avanco(x)
        correto += int(previsao == y)
    return correto / len(todos_os_y)
```

```python
# Extraindo os dados do dataset
X_treino = df[['temperatura', 'experiencia']].values
y_treino = df['classe'].values

# Inicializar o modelo
modelo = Perceptron(n_atributos=2)

# Treinar o modelo
treinamento(modelo, X_treino, y_treino, epocas=10)

# Avaliação do modelo
acuracia = computar_acuracia(modelo, X_treino, y_treino)
print(f'A acurácia do treinamento foi de {acuracia * 100}%')

# Perguntar se o usuário deseja carregar um novo arquivo CSV
resposta = input("Deseja carregar um novo arquivo CSV? (s/n): ").strip().lower()

if resposta == 's':
    df = carregar_csv()
    # Exibir as primeiras linhas do novo dataset
    print(df.head())

    # Repetir o treinamento e avaliação com o novo dataset
    X_treino = df[['temperatura', 'experiencia']].values
    y_treino = df['classe'].values

    modelo = Perceptron(n_atributos=2)
    treinamento(modelo, X_treino, y_treino, epocas=10)
    acuracia = computar_acuracia(modelo, X_treino, y_treino)
    print(f'A acurácia do treinamento foi de {acuracia * 100}%')
```
